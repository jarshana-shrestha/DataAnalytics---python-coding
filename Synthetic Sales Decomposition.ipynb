```python
# Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create synthetic date index and sales data
date_index = pd.date_range(start='2023-01-01', periods=100, freq='M')  # Monthly data
sales = np.random.randint(100, 500, size=(100))   # random sales numbers
```

```python
# Create dataframe
data = pd.DataFrame({'Date': date_index, 'Sales': sales})
data
```

```python
data.set_index('Date', inplace = True)
```

```python
# Plot
plt.figure(figsize=(12, 6)) 
plt.plot(data['Sales']) 
plt.title('Synthetic Time Series Data')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

______________

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(data['Sales'], model='additive', period=12)  # 12 months in a year

# Plot decomposition results
result.plot()
```
