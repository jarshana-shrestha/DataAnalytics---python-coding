```python
# Import necessary libraries
import matplotlib.pyplot 
import numpy as np
```

```python
# create sample data: x is an array of 100 values between 0 and 10
x= np.linspace(0, 10, 100) # returns an array of 100 evenly spaced values between 0 and 10 (inclusive)

# y is the sine of x
y = np.sin(x) # returns an array where each element is the sine of corresponsing element in x)
y2 = np.cos (x)
```

```python
# create the line plot
plt.plot (x, y, label = 'Sine Wave', color = 'black')
plt.plot (x, y2, label = 'coSine Wave', color = 'red')

# add title to the plot
plt.title('Line Plot - Sine Wave')

#Label the x- axis
plt.xlabel('X-axis')

#Label the y- axis
plt.ylabel('Y-axis')

#for the legend
plt.legend()

#Add grid
plt.grid(True)

# show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set random seed for reproducibility
np.random.seed(10)

# Create DataFrame with Age and Salary
data = pd.DataFrame({
    'Age': np.random.randint(20, 60, 100),  # Random ages from 20 to 60
    'Salary': np.random.randint(20000, 500000, 100)  # Random salaries between 20K and 500K, no. of salary 100
})

# Create a figure with 2 rows and 2 columns
plt.figure(figsize=(10, 6))  

# 1st Plot - Scatter Plot
plt.subplot(2, 2, 1)  # (rows, columns, position)
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')
plt.title('Age vs Salary (Scatter Plot)')
plt.xlabel('Age')
plt.ylabel('Salary')

# 2nd Plot - Age Histogram
plt.subplot(2, 2, 2)
sns.histplot(data['Age'], bins=5, kde=True, color='green')
plt.title('Age Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')

# 3rd Plot - Box Plot of Salary
plt.subplot(2, 2, 3)
sns.boxplot(y=data['Salary'], color='orange')
plt.title('Box Plot of Salary')
plt.ylabel('Salary')

# 4th Plot - Salary Histogram
plt.subplot(2, 2, 4)
sns.histplot(data['Salary'], bins=5, kde=True, color='blue')
plt.title('Salary Histogram')
plt.xlabel('Salary')
plt.ylabel('Frequency')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
