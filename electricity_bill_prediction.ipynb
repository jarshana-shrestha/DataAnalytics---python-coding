### Predicting Monthly Electricity Bill

 You work as a data analyst for an energy consulting firm. The firm wants to help households predict their monthly electricity bill based on the 
number of hours they use electricity each day on average. This will allow homeowners to manage their energy use more effectively. Using historical data 
collected from various households, your task is to build a simple linear regression model that predicts the monthly electricity bill based on daily
electricity usage in hours. Here is a small dataset to help you get started:

Using this data, build a linear regression model and answer the following questions:
 1. What is the expected monthly bill for a household that uses electricity for 5 hours a day?
 2. How much does the monthly bill increase, on average, for each additional hour of daily usage?
 3.  Create a visualisation that shows the data, the linear regressor, and the result value.
 
By completing this exercise, youâ€™ll better understand how linear regression can be used to predict outcomes based on input features.

```python
# Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Defining the datasets

x = np.array ([[3],[4],[6],[7],[8],[10],[12]])

y = np.array ([50, 60, 80, 95, 110, 130, 150])
```

```python
#Initializing the model

model = LinearRegression()
```

```python
# Train the model
model.fit(x, y)

# Make prediction
# Predicting the monthly bill for a household that uses electrcity for 5 hours a day
predicted_bill_5hr = model.predict(np.array([[5]])) 

print("Predicted monthly bill for 5 hours of daily usage:", predicted_bill_5hr[0]) 
```

```python
# Extract and print model parameters
#model.coef_gives the slope of the line, showing bill increase per hour
slope = model.coef_[0]

#model.intercept_ is the y-intercept, showing the starting value of the bill
intercept = model.intercept_

print("Slope (increase per additional hour):", slope)
print("Intercept (base bill):", intercept)
```

```python
# Plotting the data points
# Scatter plot for actual data points
plt.scatter(x, y, color='blue', label='Data Points')  
```

```python
# Plotting the data points
# Scatter plot for actual data points
plt.scatter(x, y, color='blue', label='Data Points')  

# Creating predictions for the entire range of X values
x_range = np.array([[x] for x in range(2, 14)])  # Generate X values for the range of interest
y_pred = model.predict(x_range)                   # Get predictionsfor the range of X values

# Plotting the regression line
plt.plot(x_range, y_pred, color='red', label='Regression Line')  # for the line

# Highlight the predicted value for 5 hours of usage
plt.scatter(5, predicted_bill_5hr, color= 'green',
label = 'Prediction for 5 hours', s = 200, edgecolor = 'black')

# Adding labels and title
plt.title('Monthly Electricity Bill Prediction')
plt.xlabel('Average Daily Usage (hours)')
plt.ylabel('Monthly Bill (dollars)')
plt.legend()                                               
plt.grid(True)

# Show the plot
plt.show()  
```
