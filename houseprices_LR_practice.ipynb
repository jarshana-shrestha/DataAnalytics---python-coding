```python
 # Import necessary libraries
 import numpy as np
 import pandas as pd
 import matplotlib.pyplot as plt
 from sklearn.model_selection import train_test_split
 from sklearn.linear_model  import LinearRegression
```

```python
# Create synthetic dataset by random 

np.random.seed(0)
house_sizes = np.random.randint(200, 2500, size=100)

print(house_sizes[:5]) 
```

```python
# Get house prices with some noise
house_prices = 50 * house_sizes + np.random.normal(0, 10000, size=100)

print(house_prices[:5])
```

```python
# Step 2: Create a DataFrame
data = pd.DataFrame({
    'Size': house_sizes,
    'Price': house_prices
 })
```

```python
# Step 3: Split the dataset into training and testing sets
x = data[['Size']]  # Features (independent variable)
y = data['Price']   # Target (dependent variable)


# 80% of the data for training and 20% for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, 
test_size=0.2, random_state=0)


# Step 4: Build and train the linear regression model
model = LinearRegression()  # Create an instance of LinearRegression
model.fit(x_train, y_train)  # Train the model

LinearRegression()
```

```python
# Step 5: Make predictions on the test set
predictions = model.predict(x_test)

# Step 6: Visualize the results
plt.figure(figsize=(10, 6))

# Plot training data
plt.scatter(x_train, y_train, color='blue', label='Training Data')

# Plot testing data
plt.scatter(x_test, y_test, color='green', label='Testing Data')

# Plot regression line
plt.plot(x_test, predictions, color='red', linewidth=2, 
label='Regression Line')

# Add labels and title
plt.title('House Price Prediction')
plt.xlabel('Size (Square Feet)')
plt.ylabel('Price ($)')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
